```python
import numpy as np
```

```python
with open('data/breast-cancer-wisconsin.data', 'rt') as f:
    temp = f.read().replace('?', '5').split()
```

```python
temp1 = temp1 = []
data = []
label = []
```

```python
for i in temp:
    temp1.append(i.split(','))
```

```python
for l in temp1:
    label.append(int(int(l[-1])/2-1))
    j = []
    for i in l:
        j.append(int(i))
    data.append(j[1:-1])
```

```python
len(data)
```

```python
T = int(len(data)*0.8)
data_train = data[:T]
data_test = data[T:]
label_train = label[:T]
label_test = label[T:]
```

```python
print('Number of samples(train, test): {}, {}'.format(len(data_train), len(data_test)))
print('Number of features: {}'.format(len(data_train[0])))
print('Number of classes: 2')
print('Valid range of feature value: 10')
```

```python
n = len(data_train)
```

```python
X = np.matrix(data_train)
Y = np.matrix(label_train).transpose()
```

```python
W1 = np.random.rand(9,5)
W2 = np.random.rand(5,1)
L = 1
learning_rate = 0.08
lossless = []
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
for i in range(10000):
    Z1 = X.dot(W1)
    L1 = sigmoid(Z1)
    Z2 = L1.dot(W2)
    y_hat = sigmoid(Z2)
    dW1 = X.transpose().dot(np.multiply(np.multiply(y_hat-Y,y_hat,1-y_hat).dot(W2.transpose()),L1,1-L1))/n
    dW2 = L1.transpose().dot(np.multiply(y_hat-Y,y_hat,1-y_hat))/n
    W1 = W1 - learning_rate * dW1
    W2 = W2 - learning_rate * dW2
    L = np.multiply(y_hat-Y,y_hat-Y).sum()/(2*n)
    lossless.append(L)
    print(L)
```

```python
from matplotlib import pyplot as plt
plt.plot(lossless)
plt.show()
```

```python
X_test = np.array(data_test)
```

```python
a = sigmoid(sigmoid(X_test.dot(W1)).dot(W2))
```

```python
y1 = []
for i in a:
    if i < 0.5:
        y1.append(0)
    else:
        y1.append(1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(label_test, y1)
```

```python
count = 0
for i in range(len(label_test)):
    if y1[i] == label_test[i]:
        count += 1
```
